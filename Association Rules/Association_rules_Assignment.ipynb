```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("/content/Online retail.xlsx")
df
```

```python
df.head()
```

```python
df.isna().sum()
df = df.drop_duplicates().dropna()
```

```python
transactions = df.iloc[:, 0].apply(lambda x: x.split(','))
```

```python
transactions.head()
```

```python
te = TransactionEncoder()

# Fit and transform the transactions into a one-hot encoded format
te_ary = te.fit_transform(transactions)

# Convert the array into a DataFrame
df = pd.DataFrame(te_ary,columns=te.columns_)

# Display the first few rows of the one-hot encoded DataFrame
df.head()
```

```python
frequent_itemsets=apriori(df,min_support=0.02,use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.5)
rules
```

```python
rules
```

```python
#Answer 1 == Lift helps to identify and quantify the strength of the relationship between items in a rule, beyond just how often they occur together
#Answer 2 == a.Support measures how frequently the itemset appears in the dataset.
#            b.Confidence measures how often items in the consequent B appear in transactions that contain the antecedent A .
#Answer 3 == Limitations and Challenges:

# a.Scalability:Large datasets can make the computation of frequent itemsets and association rules computationally expensive and time-consuming.

# b.Choice of Metrics:Selecting appropriate thresholds for support, confidence, and lift can be challenging. Setting thresholds too high might miss important rules, while too low thresholds might produce too many insignificant rules.
```
